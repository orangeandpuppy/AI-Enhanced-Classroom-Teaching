# 智慧课堂AI：AI助力课堂教学

## 1. 项目概述

### 1.1 产品名称与功能：

**名称：** 智慧课堂AI  
**功能：** 辅助教学，提升课堂效率，提供学生答题反馈  
**简介：** 基于Chatgpt和远程数据库的教育类应用，主要应用于课堂教学，提供教师端和学生端，教师端可以发布题目，查看学生总体答题情况，学生端可以在多台设备上答题，同时访问和修改数据库，获取自己的答题反馈  


### 1.2 项目背景：

去年，世界经济论坛 (WEF) 发现，教师每周投入的时间往往超过 50 个小时，而与学生的直接互动只占这项工作的不到一半。梯度学习研究显示，学生们还面临着参与度差距，50% 的学生表示，他们「大部分时间都没有投入到在学校学习的内容中」。人工智能可以通过从个性化学习之旅到预测分析等创新对教师和学生产生影响。这种技术可以帮助教师更好地了解学生的需求，提供更好的教学体验，同时也可以帮助学生更好地学习。  
在这种背景下，我们提出了智慧课堂AI这一项目，希望通过人工智能技术，提升课堂效率，提供学生答题反馈，提高学生参与度，提高教学质量。

### 1.3 项目动机：  
- 对于某些主观性较强的课程，比如中国精神导引，往往有一长段文字对应于一道题目，教师难以在课堂上快速得出得分情况，且难以公平地对待每个学生的答案。  

- 主观性较强的题目往往有多重答案，如果学生只能看到自己的回答拿到的分数，很难明白自己还有哪些知识点需要加强。  

- 通过引入大语言模型，可以在相关课程知识库中快速得到一道题的简略答案，根据此答案给每个学生的答案进行评分、写评语，并能够总结学生的总体答题情况，告诉教师学生们普遍欠缺哪些知识点的讲解。

首先，我们需要导入班级学生名单到数据库中，并确认已加载个人知识库（本项目中以中国精神导引课件为例）

```python
from util import connect_db, init_db, load_pdf
from config import config

print("a")
# 获取学生信息，并将其存入数据库
students_path = config['student_name']
student_name = []
with open(students_path, 'r', encoding='utf-8') as f:
    lines = f.readlines()
    student_name = [line.strip() for line in lines]
with connect_db() as con:
    cur = con.cursor()
    # 判断数据库charging_db是否存在，不存在则创建
    if cur.execute("SHOW DATABASES LIKE 'charging_db';") == 0:
        init_db(con)
        cur.execute("USE charging_db;")
        cur.execute("UPDATE class SET student_count = %d WHERE class_id = 1;" % (len(student_name),))
        for name in student_name:
            cur.execute("INSERT INTO students VALUES ('%s');" % (name,))
        con.commit()
    print("班级学生已存入数据库！")

# 加载PDF文档
with connect_db() as con:
    cur = con.cursor()
    cur.execute("USE charging_db;")
    cur.execute("SELECT len FROM pdflen WHERE id = 1;")
    len = cur.fetchone()[0]
    print("数据库中已有文档长度：", len)
    if len == 0:
        load_pdf()
        print("PDF文档已加载！")
    else:
        print("PDF文档已加载！")
```

接下来，我们测试教师端功能，发布一道题目，并查看学生提交的答案情况。
为了方便测试，建议选择发布第一道题目，因为数据库中已经存储了几十个学生第一道题目的答案，可以查看学生总体答题情况。

```python
from config import config
from util import connect_db, get_standard_answer, get_overall_situation
from wordcloud import WordCloud
import matplotlib.pyplot as plt


class Teacher:
    def __init__(self):
        pass

    def publish_question(self):
        """
            发布题目
        """
        # 检查是否有题目没有终止答题
        with connect_db() as con:
            cur = con.cursor()
            cur.execute("USE charging_db;")
            cur.execute("SELECT * FROM questions;")
            if cur.rowcount > 0:
                print("有题目未终止答题！")
                return

        # 选择题目
        print("题目列表如下：\n")
        with open(config['question'], 'r', encoding='utf-8') as file:
            questions = [line.strip() for line in file]
        for i, question in enumerate(questions, 1):
            print(f"{i}. {question}")
        question_id = int(input("\n请选择要发布的题目编号："))
        question = questions[question_id - 1]

        standard_answer = get_standard_answer(question)

        # 发布题目
        with connect_db() as con:
            cur = con.cursor()
            cur.execute("USE charging_db;")
            cur.execute("INSERT INTO questions (question_id, question, standard_answer) VALUES (%d , '%s', '%s');"
                        % (question_id, question, standard_answer))
            con.commit()
            print(f"题目 {question} 发布成功！")

    def view_current_situation(self):
        """
            查看目前答题情况
        """
        with connect_db() as con:
            cur = con.cursor()
            cur.execute("USE charging_db;")
            cur.execute("SELECT * FROM questions;")
            if cur.rowcount == 0:
                print("暂无题目正在答题！")
            else:
                question = cur.fetchone()
                question_id = question[0]
                question = question[1]
                print(f"当前正在答题的题目：{question_id}. {question}")
                # 统计学生总数
                cur.execute("SELECT * FROM students;")
                student_count = len(cur.fetchall())
                # 统计已经提交答案的学生人数
                cur.execute("SELECT * FROM answers WHERE question_id = %d;" % question_id)
                answered_count = cur.rowcount
                print(f"该题目已有 {answered_count} 人提交答案，共 {student_count} 人。")

    def stop_answer(self):
        """
            终止答题
        """
        question_id = 1
        question = ""
        standard_answer = ""
        with connect_db() as con:
            cur = con.cursor()
            cur.execute("USE charging_db;")
            cur.execute("SELECT * FROM questions;")
            if cur.rowcount == 0:
                print("暂无题目正在答题！")
                return -1
            else:
                question = cur.fetchone()
                question_id, question, standard_answer = question
                print(f"当前正在答题的题目：{question_id}. {question}")
                confirm_end = input("是否终止题目 %d 的答题？(Y/N)\n".format(question_id)).upper()
                while confirm_end != 'Y' and confirm_end != 'N' and confirm_end != 'YES' and confirm_end != 'NO':
                    print("指令不能识别，请重新输入！")
                    confirm_end = input("是否终止题目 %d 的答题？(Y/N)\n".format(question_id)).upper()
                if confirm_end == 'Y' or confirm_end == 'YES':
                    cur.execute("DELETE FROM questions;")
                    print("题目 %s 的答题已终止！" % question)
                    con.commit()
                else:
                    print("操作已取消！")
                    return -1

        # 得到总体答题情况
        get_overall_situation(question_id, question, standard_answer)
        return question_id

    def view_overall_situation(self, question_id: int):
        """
            查看总体答题情况
        """
        with connect_db() as con:
            cur = con.cursor()
            cur.execute("USE charging_db;")
            cur.execute("SELECT * FROM overall_situations WHERE question_id = %d;" % question_id)
            if cur.rowcount == 0:
                print(f"暂无题目 {question_id} 的总体答题情况！")
            else:
                overall_situation = cur.fetchone()
                # 用,分割每一个关键词及其频率
                keywords_freq_list = overall_situation[1].split(',')
                keywords_freq = dict()
                for keyword_freq in keywords_freq_list:
                    keyword, freq = keyword_freq.split(':')
                    keywords_freq[keyword] = float(freq)

                # 创建词云对象
                wordcloud = WordCloud(
                    width=800,
                    height=800,
                    background_color='white',
                    font_path='font/msyh.ttc',
                    max_words=200,
                    max_font_size=100,
                    random_state=42
                ).generate_from_frequencies(keywords_freq)

                # 显示词云图
                plt.figure(figsize=(10, 10))
                plt.imshow(wordcloud, interpolation='bilinear')
                plt.axis('off')
                plt.show()

                print(f"问题 {question_id} 的总体答题情况：")
                analysis = overall_situation[2]
                print(analysis)

teacher = Teacher()
teacher.publish_question()
```

查看有多少学生提交了答案

接下来，我们测试学生端功能，提交正在进行的题目的答案，并查看自己的答题反馈。  
学生姓名必须是班级学生名单中的姓名，且只能提交一次答案。比如“张三”是一个可以提交答案的学生。

```python
from util import connect_db, get_feedback


class Student:
    def __init__(self, student_name):
        self.student_name = student_name

    # 提交答案
    def submit_answer(self):
        question_id = 1
        question = ""
        standard_answer = ""

        # 读取发布的题目
        with connect_db() as con:
            cur = con.cursor()
            cur.execute("USE charging_db;")
            cur.execute("SELECT * FROM questions;")
            questions = cur.fetchall()
            if len(questions) == 0:
                print("暂无题目！")
                return
            for question in questions:
                question_id, question, standard_answer = question
            print(f"发布的题目：{question}")

        # 检查是否已经提交过答案
        with connect_db() as con:
            cur = con.cursor()
            cur.execute("USE charging_db;")
            cur.execute("SELECT * FROM answers WHERE student_name = '%s' AND question_id = %d;"
                        % (self.student_name, question_id))
            if cur.rowcount > 0:
                print("您已经提交过答案！")
                return

        # 输入答案
        answer = input("请输入答案：")
        with connect_db() as con:
            cur = con.cursor()
            cur.execute("USE charging_db;")
            cur.execute("INSERT INTO answers VALUES ('%s', '%s', %d);" % (self.student_name, answer, question_id))
            con.commit()
            print("答案已提交！\n")

        # 将学生的答案传递给大语言模型，获取反馈
        score, feedback = get_feedback(self.student_name, answer, standard_answer, question_id, question)
        print(f"得分：{score}")
        print(f"反馈：{feedback}")

        return question_id

    # 查询反馈
    def query_feedback(self, question_id):
        with connect_db() as con:
            cur = con.cursor()
            cur.execute("USE charging_db;")
            cur.execute("SELECT * FROM feedbacks WHERE student_name = '%s' AND question_id = %d;" % (self.student_name, question_id))
            if cur.rowcount == 0:
                print(f"暂无问题 {question_id} 的反馈！")
            else:
                feedback = cur.fetchone()
                print(f"问题 {question_id} 的反馈：")
                print(f"得分：{feedback[1]}")
                print(f"反馈：{feedback[2]}")
                print("\n")


def in_db(student_name):
    with connect_db() as con:
        cur = con.cursor()
        cur.execute("USE charging_db;")
        cur.execute("SELECT * FROM students WHERE name = '%s';" % (student_name,))
        return cur.fetchone() is not None

# 检查数据库是否存在
with connect_db() as con:
    cur = con.cursor()
    if cur.execute("SHOW DATABASES LIKE 'charging_db';") == 0:
        raise Exception("数据库不存在！")
    
# 输入学生姓名
student_name = input("请输入学生姓名：")
while not in_db(student_name):
    student_name = input("学生姓名不存在，请重新输入：")
print("成功登陆！\n")

student = Student(student_name)
student.submit_answer()
```
